```python
x=iris.iloc[:,:4]
x.head()
```

```python
y=iris.iloc[:,-1]
y.head()
```

```python
from sklearn import preprocessing
x=preprocessing.StandardScaler().fit_transform(x)
x[0:4]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
y_test.shape
```

```python
#training and predicting
from sklearn.neighbors import KNeighborsClassifier
knnmodel=KNeighborsClassifier(n_neighbors=3)
```

```python
knnmodel.fit(x_train,y_train)
```

```python
KNeighborsClassifier(algorithm='auto',leaf_size=30,metric='minkowski',metric_params=None,n_jobs=None,n_neighbors=3,p=2,weights='uniform')
```

```python
y_predict1=knnmodel.predict(x_test)
```

```python
#Accuracy

from sklearn.metrics import accuracy_score
```

```python
acc=accuracy_score(y_test,y_predict1)
acc               
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test.values,y_predict1)
cm
```

```python
cm1=pd.DataFrame(data=cm,index=['setosa','versicolor','verginica'],columns=['setosa','versicolor','verginica'])
cm1
```

```python
#output visualization
prediction_output=pd.DataFrame(data=[y_test.values,y_predict1],index=['y_test','y_predict1'])
```
